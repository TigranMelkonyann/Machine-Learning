```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Generate linear data with two classes
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, 
                           n_redundant=0, n_classes=2, n_clusters_per_class=1, 
                           random_state=41, hypercube=False, class_sep=11)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.title("Generated Data with Two Classes")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
lor = LogisticRegression(penalty=None, solver='sag')
lor.fit(X_train, y_train)
```

```python
print("Coefficients:", lor.coef_)
print("Intercept:", lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0] / lor.coef_[0][1])  # Slope of the decision boundary
b1 = -(lor.intercept_ / lor.coef_[0][1])   # Intercept of the decision boundary
x_input = np.linspace(-3, 3, 100)          
y_input = m1 * x_input + b1                
```

```python
# Visualize the decision boundary and training data
plt.figure(figsize=(7, 5))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='winter', s=100, label="Training Data")
plt.plot(x_input, y_input, color='red', linewidth=3, label="Decision Boundary")
plt.title("Logistic Regression Decision Boundary (Training Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.ylim(-3, 2)
plt.show()
```

```python
# Visualize the decision boundary and testing data
plt.figure(figsize=(7, 5))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='winter', s=100, label="Testing Data")
plt.plot(x_input, y_input, color='red', linewidth=3, label="Decision Boundary")
plt.title("Logistic Regression Decision Boundary (Testing Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.ylim(-3, 2)
plt.show()
```
